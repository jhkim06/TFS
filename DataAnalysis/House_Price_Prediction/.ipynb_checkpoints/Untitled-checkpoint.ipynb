```python
sns.distplot(train['SalePrice'])   # 판매가의 히스토그램 확인
```

```python
# log를 취한 값으로 바꾼다.
# np.log1p는 밑은 자연상수 e, x에 대한 로그를 취할때 x+1 값을 대입한 결과를 출력함
train["SalePrice"] = np.log1p(train["SalePrice"])
sns.distplot(train['SalePrice']);
```

```python
ntrain = train.shape[0]   # 학습데이터 갯수: 1460개
ntest = test.shape[0]   # 시험데이터 갯수: 1459개 
y_train = train.SalePrice.values   # .values가 붙을 시 np.array로 전환됨
```

```python
y_train
```

```python
# train data와 test data를 합친다.
all_data = pd.concat((train, test), sort=False).reset_index(drop=True)
all_data.drop(['SalePrice'], axis=1, inplace=True)   # inplace=True로 해야 drop된 결과가 all_data변수에 반영된다.
all_data.shape
```

```python
all_data.loc[1457:1462]   # trainset과 testset의 접합부위 확인
```

```python
all_data.isnull().head(2)
```

```python
# 결측치를 모두 확인 
all_na = (all_data.isnull().sum() / len(all_data)) * 100

# 결측치가 없는 열들은 제외한다.
all_na = all_na.drop(all_na[all_na == 0].index).sort_values(ascending=False)[:34] # 결측치가 많은 상위 30개만 골라냄
missing_data = pd.DataFrame({'Missing Ratio' : all_na})  # DataFrame 생성방법
missing_data.head(34)
```

```python
# 결측치 비율을 막대그래프로 나타냄
f, ax = plt.subplots(figsize=(15, 12))
plt.xticks(rotation='90')
sns.barplot(x=all_na.index, y=all_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
```
