```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
train_data=pd.read_csv('../Data/Titanic/trainset.csv')
test_data=pd.read_csv('../Data/Titanic/testset.csv')
combine = [train_data, test_data]
print(combine)
```

```python
train = train_data[:750]   # Validation Set의 분리
vld = train_data[750:]
```

```python
x_train = train.drop("Survived", axis=1)
y_train = train["Survived"]
```

```python
x_train.tail(2)
```

```python
y_train.tail(2)
```

```python
x_train = x_train.values   # DataFrame에서 index와 header를 제거하고 수치부분만 떼어내어 numpy.array형태로 변환
x_train = x_train.astype('float')   # numpy.array에서 자료형을 실수형으로 전환
print(x_train)
```

```python
y_train = y_train.values   # y_train에서도 수치부분만 떼어냄
y_train = y_train.reshape(750,1)   # Label의 shape를 재조정
y_train = y_train.astype('float')  # Label의 자료형을 실수형으로 전환
print(y_train)
```

```python
x_vld = vld.drop('Survived', axis = 1)   # Validation Set의 Feature vs Label 분리
y_vld = vld['Survived']
```

```python
x_vld.tail(2)
```

```python
x_vld = x_vld.values
x_vld = x_vld.astype('float')
```

```python
print(x_vld)
```

```python
y_vld.tail(2)
```

```python
y_vld = y_vld.values
y_vld = y_vld.reshape(141,1)
y_vld = y_vld.astype('float')
```

```python
print(y_vld)
```

```python
print(x_train.shape, y_train.shape)
```

```python
model = tf.keras.models.Sequential([
  # tf.keras.layers.Flatten(input_shape=(28, 28)),   # Flatten을 적용할 필요 없음 (MNIST와의 차이)
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(16, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)   # 학습
```

```python
model.evaluate(x_vld, y_vld)   # Validation Set을 통한 정확도 확인
```

```python
X_train = train_data.drop('Survived', axis=1)   # Validation Set 도 포함하여 학습
X_train = X_train.values
X_train = X_train.astype('float')
Y_train = train_data["Survived"]
Y_train = Y_train.values
Y_train = Y_train.reshape(891,1)
Y_train = Y_train.astype('float')
```

```python
print(X_train.shape, Y_train.shape)
```

```python
model.fit(X_train, Y_train, epochs=10)
```

```python
test = test_data.drop('PassengerId', axis = 1)
test = test.values
test = test.astype('float')
print(test, test.shape)
```

```python
Submission = pd.DataFrame(test_data['PassengerId'])   # PassengerId 하나만 열로 갖는 데이터프레임 생성
test_predict = model.predict(test)   # test 대상에 대해 생존여부 예측

# One-hot encoding되어 출력되는 결과를 decoding해주는 부분
test_predict = test_predict.tolist()
for i,j in enumerate(test_predict):
    if j[0] > j[1]: test_predict[i] = 0
    else: test_predict[i] = 1

Submission["Survived"] = test_predict
print(Submission)
```

```python
Submission.to_csv("../Data/Titanic/tf_summission.csv", header = True, index = False)   # 저장
```
