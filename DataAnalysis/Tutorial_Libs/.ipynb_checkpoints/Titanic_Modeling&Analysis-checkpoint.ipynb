# 타이타닉 실습_분석 모델링 

```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
sklearn.__version__   # sklearn 버젼 확인
```

```python
trainset = pd.read_csv('../Data/Titanic/trainset.csv')
testset = pd.read_csv('../Data/Titanic/testset.csv')
combine = [trainset, testset]
combine
```

```python
trainset.head()
```

```python
testset.tail()
```

```python
print(trainset.shape, testset.shape)
```

```python
train = trainset.loc[:750]   # Validation Set 의 분리
vld = trainset.loc[750:]
```

```python
train.tail(2)
```

```python
vld.tail(2)
```

```python
X_train = train.drop('Survived', axis = 1)   # axis=1 은 "Survived" 컬럼을 없앤다는 뜻
Y_train = train['Survived']   # Label 의 분리
X_vld = vld.drop('Survived', axis = 1)   # Validation Set도 마찬가지로 처리
Y_vld = vld['Survived']
```

```python
X_train.tail(2)
```

```python
Y_train.tail(2)
```

```python
X_vld.tail(2)
```

```python
Y_vld.tail(2)
```

```python
lr = LogisticRegression(solver='lbfgs')
lr.fit(X_train, Y_train)
```

```python
lr.score(X_vld, Y_vld)   # LogisticRegression 방법의 정확도
```

```python
acc_lr = round(lr.score(X_vld,Y_vld)*100, 4)
acc_lr
```

```python
svc = SVC(gamma='scale')   # 서포트 벡터 머신의 이용
svc.fit(X_train, Y_train)
acc_svc = round(svc.score(X_vld,Y_vld)*100, 4)
acc_svc
```

```python
dt = DecisionTreeClassifier()   # Decision Tree 방법의 이용
dt.fit(X_train, Y_train)
dt_svc = round(dt.score(X_vld,Y_vld)*100, 4)
dt_svc
```

```python
rf = RandomForestClassifier(n_estimators=100)   # Random Forest 방법의 이용
rf.fit(X_train, Y_train)
rf_svc = round(rf.score(X_vld,Y_vld)*100, 4)
rf_svc
```

```python
X_Trainset = trainset.drop('Survived', axis = 1)   # Validation Set을 포함해서 학습!!
Y_Trainset = trainset['Survived']
```

```python
svc.fit(X_Trainset, Y_Trainset)   # SVC의 정확도가 가장 높으므로 이 방법을 채택!
```

```python
X_Testset = testset.drop('PassengerId', axis = 1)   # 생존여부의 예측대상인 testset에서 승객번호 열을 제거
X_Testset.tail()
```

```python
Y_pred = svc.predict(X_Testset)   # 예측대상에 대한 생존여부의 예측
Y_pred
```

```python
submission_df = pd.DataFrame(testset['PassengerId'])   # PassengerId Series를 DataFrame으로 변환
submission_df.head()
```

```python
submission_df['Survived'] = Y_pred   # 앞서 구한 생존여부 예측결과를 새로운 컬럼으로 추가
submission_df
```

```python
submission_df.to_csv("../Data/Titanic/submission.csv", header = True, index = False)   # 저장
```
