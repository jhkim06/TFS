```python
# 전처리하기 여기서는 "|" 를  "," 로 바꿔준다
def make_csv_format(csv_raw_path=DATA_PATH, csv_raw_name=None):
    if not os.path.exists(csv_raw_path+'/new'):
        os.makedirs(csv_raw_path+'/new')
    
    new_csv = open(csv_raw_path+'/new/'+ 'new_' + csv_raw_name,'w')
    with open(csv_raw_path+'/'+csv_raw_name,'r') as filelist:
        for afile in filelist:
            #afile = (afile.rstrip('\n')).replace('|',',')
            afile = afile.replace('|',',')
            new_csv.write(afile)
            #print(afile)
    new_csv.close()
```

```python
import geopandas as gpd
import descartes

SHAPE_DIR = "/Volumes/Samsung_T3/SKT_1/DMM_BLOCK/"
SHAPE_FILE = ['DMM_BLOCK.SHX']

data = gpd.read_file(SHAPE_DIR + SHAPE_FILE[0])

# 원하는 시, 구에 해당하는 지도 선택
def select_SI_DO_GU(data_raw, si, gu, adong = None, ldong = None):
    selected_map = data.loc[data_raw["SIDO_NM"] == si]
    selected_map = selected_map.loc[data["SGNG_NM"] == gu]
    if not adong == None:
        selected_map = selected_map.loc[data["ADONG_NM"] == adong]
    
    if not ldong == None:
        selected_map = selected_map.loc[data["LDONG_NM"] == ldong]
        
    return selected_map
    
# 특정 블락 아이디가 선택한 영역에 있는지 확인
def scan_block_ids(id_, ids_blocks):
    for id_block in ids_blocks:
        if id_block == id_:
            return True
    return False 

# 데이터에서 선택한 지도 영역에 속하는 블락 아이디만 고르기
def select_blocks(data_, ids_blocks):
    blocks = data_["BLOCK_CD"]
    in_test_set = blocks.apply(lambda id_: scan_block_ids(id_, ids_blocks))
    return data_.loc[in_test_set]
```

```python
time_name_dict={}
for time in time_name:
    time_name_dict[time] = time.split("_")[1] + "시"
```

```python
#
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl

def draw_map_and_scatter(scatter_data, map_data, si, gu, adong = None, ldong = None, color_value = None, color_value_name = None):
    
    cmap = plt.cm.jet  # define the colormap
    # extract all colors from the .jet map
    cmaplist = [cmap(i) for i in range(cmap.N)]
    # force the first color entry to be grey
    cmaplist[0] = (.5, .5, .5, 1.0)
    
    # create the new map
    cmap = mpl.colors.LinearSegmentedColormap.from_list(
    'Custom cmap', cmaplist, cmap.N)
    
    #tick_wratio = np.linspace(0, 300, 11)
    tick_wratio = [1, 10, 50, 100, 500, 1000, 2000, 3000]
    print (tick_wratio)
    norm = mpl.colors.BoundaryNorm(tick_wratio, cmap.N)

    seoul_map = select_SI_DO_GU(map_data, si, gu, adong, ldong)
    total_data_seoul = select_blocks(scatter_data, seoul_map["BLOCK_CD"])
    
    fig, ax = plt.subplots()
    fig.set_size_inches(10,10)
    
    plt.subplots_adjust(bottom=0.1, right=0.3, top=0.3)

    path=ax.scatter(x = total_data_seoul["X_COORD"], y = total_data_seoul["Y_COORD"], 
                    s = total_data_seoul["total_population"]/20, label="population",
                    c = total_data_seoul[color_value], cmap=cmap, norm=norm, alpha = 0.4
                   )
    
    seoul_map.plot(ax=ax, color='white', edgecolor='black', alpha=0.5)
    
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    
    # 컬러바 그리기
    cbar = plt.colorbar(path, cax=cax, ticks=tick_wratio, norm = norm, cmap=cmap, spacing='uniform', 
                        boundaries=tick_wratio )
    
    si_ = si
    gu_ = gu
    adong_ = adong
    ldong_ = ldong
    
    # 제목으로 지역정보 나타내기 (시군구동) 
    a = lambda s: '' if s is None else str(s)
    temp_title = a(si_) + " " + a(gu_) + " " + a(adong_) + " " + a(ldong_)
    fig.suptitle(temp_title, fontproperties=fontprop)
    
    # 레전드(원) 사이즈 조절
    #lgnd = plt.legend(scatterpoints=1, fontsize=10)
    #lgnd.legendHandles[0]._sizes = [20]

    #cax = fig.get_axes()[1] 
    cax.set_ylabel(time_name_dict[color_value] + " " + color_value_name, fontproperties=fontprop)

    save_fig(si+gu+"_"+color_value+color_value_name+"_visualization_plot")
    
    del fig, ax, cax, divider, seoul_map, total_data_seoul,  tick_wratio
```

```python
# 테스트 해보기
draw_map_and_scatter(total_data, data, "서울특별시", "종로구", None, None, "TMST_00", "일평균 유동인구 수 (명)")
```
